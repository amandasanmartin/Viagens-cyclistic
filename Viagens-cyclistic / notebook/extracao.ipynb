```python
import requests
import zipfile
import io
import os

# 1. Importação para a pasta
caminho_pasta = r'C:\Projeto_Cyclistic'
os.makedirs(caminho_pasta, exist_ok=True)

# 2. Define os anos e meses para o loop
anos = ["2021", "2022", "2023", "2024", "2025"]
meses = [f"{i:02d}" for i in range(1, 13)] # Gera '01' a '12'

print("Iniciando o processo de extração...\n")

for ano in anos:
    for mes in meses:
        # URL padrão da base de dados da Divvy Bikes
        url = f"https://divvy-tripdata.s3.amazonaws.com/{ano}{mes}-divvy-tripdata.zip"
        print(f"Buscando: {ano}-{mes}...")
        
        resposta = requests.get(url)
        
        if resposta.status_code == 200:
            # Extrai o conteúdo do ZIP direto para a local C:\
            with zipfile.ZipFile(io.BytesIO(resposta.content)) as z:
                z.extractall(caminho_pasta)
            print(f"  -> {ano}-{mes} extraído com sucesso!")
        else:
            print(f"  -> Atenção: Arquivo de {ano}-{mes} não encontrado (pode ter outro formato de nome).")

print("\nDownload e extração finalizados!")
```

```python

import pandas as pd
import os
# Extração das tabelas 
caminho_base = r'C:\Projeto_Cyclistic'
nome_tabela = 'viagens_cyclistic'
anos = ["2021", "2022", "2023", "2024", "2025"]
meses = [f"{i:02d}" for i in range(1, 13)]

#Consolidação das tabelas
print("Iniciando a consolidação por ano...\n")

for ano in anos:
    print(f"Agrupando dados de {ano}...")
    arquivo_saida = os.path.join(caminho_base, f"{ano}_consolidado.csv")
    
    # Variável para garantir que o cabeçalho seja escrito apenas na 1ª linha do ano
    cabecalho_escrito = False
    
    for mes in meses:
        nome_arquivo = f"{ano}{mes}-divvy-tripdata.csv"
        
        # Tratando a exceção da mudança de nome feita pela equipe da Cyclistic
        if ano == "2022" and mes == "09":
            nome_arquivo = "202209-divvy-publictripdata.csv"
            
        caminho_completo = os.path.join(caminho_base, nome_arquivo)
        
        if os.path.exists(caminho_completo):
            # Leitura em blocos de 100 mil linhas para economizar memória RAM
            for chunk in pd.read_csv(caminho_completo, chunksize=100000):
                chunk.to_csv(arquivo_saida, mode='a', index=False, header=not cabecalho_escrito)
                cabecalho_escrito = True
                
    print(f"-> {ano}_consolidado.csv salvo com sucesso!\n")
```
